```python
import osmnx as ox
import os
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the bounding box
north, south, east, west = 64.188968, 64.033992, -21.640345, -22.083955

tags = {"amenity": ["parking", "parking_space", "parking_entrance"]}
parking_data = ox.geometries_from_bbox(north, south, east, west, tags)
print('Saved parking lot data to file')
```

```python
print(parking_data.head())
```

```python
parking_gdf = gpd.GeoDataFrame(parking_data, geometry='geometry')
parking_gdf.set_crs(epsg=4326, inplace=True)
```

```python
# Plot the GeoDataFrame
ax = parking_gdf.plot(figsize=(10, 10))

# Set the x and y limits
ax.set_xlim(-21.96, -21.92)
ax.set_ylim(64.11, 64.13)

# Add a title and show the plot
plt.title("Parking Lots in Restricted Area")
plt.show()
```

```python
parking_gdf.info()
```

```python
# List of selected columns
selected_columns = [
    'access', 'amenity', 'capacity', 'fee', 'parking', 'surface',
    'geometry', 'operator', 'covered', 'capacity:disabled'
]

# Filter the DataFrame
parking_gdf_filtered = parking_gdf[selected_columns]

# Check the filtered DataFrame
print(parking_gdf_filtered.head())
```

```python
# Count unique geometry types
geometry_counts = parking_gdf_filtered['geometry'].geom_type.value_counts()

# Display the results
print(geometry_counts)
```

```python
# Filter for Point geometries
point_geometries = parking_gdf_filtered[parking_gdf['geometry'].geom_type == 'Point']

# Count non-null capacity values for Point geometries
points_with_capacity = point_geometries['surface'].notnull().sum()

print(f"Number of Point geometries with capacity data: {points_with_capacity}")
```

```python
from shapely.geometry import MultiPolygon

# Filter out Point geometries
filtered_gdf = parking_gdf_filtered[parking_gdf['geometry'].geom_type != 'Point']

# Convert Polygon to MultiPolygon
filtered_gdf['geometry'] = filtered_gdf['geometry'].apply(
    lambda geom: MultiPolygon([geom]) if geom.geom_type == 'Polygon' else geom
)

# Verify the transformation
print(filtered_gdf['geometry'].geom_type.value_counts())
```

```python
# Create a copy of the GeoDataFrame to avoid overwriting original data
gdf_with_area = filtered_gdf.copy()

# Reproject to a metric CRS for accurate area calculation
gdf_with_area = gdf_with_area.to_crs(epsg=3857)

# Calculate area and add a new field
gdf_with_area['calculated_area'] = gdf_with_area['geometry'].area

gdf_with_area = gdf_with_area[gdf_with_area['calculated_area'] < 100000]

# Reproject back to the original CRS (if necessary)
gdf_with_area = gdf_with_area.to_crs(parking_gdf.crs)

# Inspect the updated GeoDataFrame
print(gdf_with_area[['geometry', 'calculated_area']].head())
```

```python
import matplotlib.pyplot as plt

# Filter the GeoDataFrame for entries with non-null capacity and calculated_area
filtered_data = gdf_with_area[gdf_with_area['capacity'].notnull(
) & gdf_with_area['calculated_area'].notnull()]

# Ensure the capacity column is numeric
filtered_data['capacity'] = pd.to_numeric(
    filtered_data['capacity'], errors='coerce')

# Drop rows where capacity conversion failed (NaN)
filtered_data = filtered_data.dropna(subset=['capacity'])

# Plot the correlation
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['calculated_area'],
            filtered_data['capacity'], alpha=0.6)
plt.title('Correlation Between Area and Capacity of Parking Lots')
plt.xlabel('Calculated Area (sq meters)')
plt.ylabel('Capacity')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Perform linear regression
x = filtered_data['calculated_area'].values
y = filtered_data['capacity'].values

# Linear regression (forcing y = ax, so intercept = 0)
slope, _, r_value, _, _ = linregress(x, y)

# Calculate how trustworthy the slope is (R-squared)
r_squared = r_value**2

# Display the results
print(f"Estimated slope (a): {slope:.4f}")
print(f"R-squared (trustworthiness): {r_squared:.4f}")

# Plot the data and regression line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, label="Data")
plt.plot(x, slope * x, color='red',
         label=f"y = {slope:.4f}x (R² = {r_squared:.4f})")
plt.title('Correlation Between Area and Capacity of Parking Lots')
plt.xlabel('Calculated Area (sq meters)')
plt.ylabel('Capacity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
gdf_with_area['estimated_capacity'] = gdf_with_area.apply(
    lambda row: row['capacity'] if pd.notnull(row['capacity']) else row['calculated_area'] * slope, axis=1
)

display(gdf_with_area)
```

```python
gdf_with_area['estimated_capacity'] = gdf_with_area['estimated_capacity'].astype(
    float)

gpd = gpd.GeoDataFrame(gdf_with_area)
gpd.to_file("../data/parking_lots.geojson", driver='GeoJSON')
display(gpd)
```
