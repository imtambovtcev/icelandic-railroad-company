```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import numpy as np
```

```python
# Load Excel file
file_path = "../data/coords.xlsx"
excel_data = pd.ExcelFile(file_path)

# Initialize an empty list to store DataFrames
dataframes = []

# Iterate through all sheets and append to the list
for sheet_name in excel_data.sheet_names:
    print(f"Processing {sheet_name}")
    # Read the sheet into a DataFrame
    df = excel_data.parse(sheet_name)
    display(df)
    df['Latitude'] = df['Latitude'].astype(float)
    df['Longitude'] = df['Longitude'].astype(float)

    # Add a category column with the sheet name
    df['category'] = sheet_name

    min_reviews = df['Reviews'].min()
    df['Reviews'].fillna(min_reviews, inplace=True)

    df['Reviews'] = df['Reviews'].astype(int)

    # Append to the list
    dataframes.append(df)


# Combine all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
coefficients = {
    "Tourism": 1.0,
    "GymsSports": 1.5,
    "Commercial Dist.": 5.0,
    "ParksStreetRecreationReligious.": 1.2,
    "Pools": 1.4,
    "Adm. Districts": 10.0,
    "EduLibHealthcare": 20.0,
    "Trans. Hub": 10.0
}

combined_df['score'] = combined_df.apply(
    lambda row: coefficients[row['category']] * row['Reviews'], axis=1)
combined_df['normalized_score'] = combined_df['score'] / \
    combined_df['score'].max()

combined_df['sigmoid_normalized_score'] = combined_df['normalized_score'].apply(
    lambda x: 1 / (1 + np.exp(-x)))
display(combined_df)
```

```python
unique_categories = combined_df['category'].unique()
print(unique_categories)
```

```python
# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(
    combined_df['Longitude'], combined_df['Latitude'])]
gdf = gpd.GeoDataFrame(combined_df, geometry=geometry)

# Set the CRS (Coordinate Reference System)
gdf.set_crs(epsg=4326, inplace=True)  # WGS84
```

```python
gdf.to_file("../data/poi_data.geojson", driver="GeoJSON")
```

```python

with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', None, 'display.max_colwidth', None):
    gdf_sorted = gdf.sort_values(by='normalized_score', ascending=False)
    display(gdf_sorted)
```
