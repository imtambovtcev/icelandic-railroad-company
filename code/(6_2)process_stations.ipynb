```python
from railroad import load_graph, save_graph
import pandas as pd
```

```python
G = load_graph('../data/reykjavik_road_graph_with_bridges.json')
```

```python
stations_df = pd.read_csv('../data/6_stations_to_years.csv')
display(stations_df)
```

```python
coord_x, coord_y = [], []

for i, row in stations_df.iterrows():
    id = row['intersection_id']
    point = G.nodes[id]['geometry']
    coord_x.append(point.x)
    coord_y.append(point.y)

stations_df['coord_x'] = coord_x
stations_df['coord_y'] = coord_y
display(stations_df)
stations_df.to_csv('../data/6_stations_to_years_with_coords.csv', index=False)
```
