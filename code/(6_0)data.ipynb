```python
import geopandas as gpd
from railroad import load_graph
from utils import print_graph_info
```

```python
# Define output directory for saving files
output_dir = "../data/"
gdfs = {}
# Process and save each GeoDataFrame
for year in [2025, 2029, 2030]:
    print(f"Processing: {year}")
    gdf = gpd.read_file(f"../datahon_shared_files/cityline_{year}.geojson")
    # Convert to WGS 84 if not already in that CRS
    if gdf.crs != "EPSG:4326":
        gdf = gdf.to_crs("EPSG:4326")
    display(gdf)
    gdfs[year] = gdf
    # Construct the output file name
    output_file = f"{output_dir}cityline_{year}_foursquare.geojson"

    # Save the GeoDataFrame as GeoJSON
    gdf.to_file(output_file, driver="GeoJSON")
    print(f"Saved: {output_file}")
```
